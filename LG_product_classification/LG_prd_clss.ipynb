##### 1. 하이퍼파라미터 최적화
 - Optuna를 활용하여 파라미터 탐색
 - 'Y_Quality' 분포는 정규분포를 따르므로 평균제곱오차, 'Y_Class'는 F1-score 평가지표 활용
 - 리소스 이슈로 회귀모델의 optuna trial 횟수는 10회로 제한하고 분류모델은 optuna 사용 안함

```python
import optuna
from optuna import Trial
from optuna.samplers import TPESampler
from sklearn.metrics import mean_squared_error
from catboost import CatBoostClassifier, CatBoostRegressor # catboost
```

```python
sampler = TPESampler(seed=1)

def objective(trial):

    cbrm_param = {
        'iterations':trial.suggest_int("iterations", 1000, 20000),
        'od_wait':trial.suggest_int('od_wait', 500, 2300),
        'learning_rate' : trial.suggest_uniform('learning_rate',0.01, 1),
        'reg_lambda': trial.suggest_uniform('reg_lambda',1e-5,100),
        'subsample': trial.suggest_uniform('subsample',0,1),
        'random_strength': trial.suggest_uniform('random_strength',10,50),
        'depth': trial.suggest_int('depth',1, 15),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf',1,30),
        'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations',1,15),
        'bagging_temperature' :trial.suggest_loguniform('bagging_temperature', 0.01, 100.00),
        'colsample_bylevel':trial.suggest_float('colsample_bylevel', 0.4, 1.0),
    }

    model_cbrm = CatBoostRegressor(**cbrm_param)
    model_cbrm = model_cbrm.fit(trimmed_train_df, train_df['Y_Quality'], eval_set=[(x_val, y_val)], 
                           verbose=0, early_stopping_rounds=25)

    MSE = mean_squared_error(y_val, model_cbrm.predict(x_val))
    return MSE

optuna_cbrm = optuna.create_study(direction='minimize', sampler=sampler)
optuna_cbrm.optimize(objective, n_trials = 10)
```

```python
# 회귀모델 하이퍼파라미터 최적화 사항 저장

import joblib

joblib.dump(optuna_cbrm, "cbr_study_tuning_model.pkl")
```

```python
cbrm_trial = optuna_cbrm.best_trial
cbrm_trial_params = cbrm_trial.params
print('Best Trial: score {},\nparams {}'.format(cbrm_trial.value, cbrm_trial_params))
```

##### 2. 회귀 모델

- CatBoostRegressor를 통한 회귀 모델
- 타겟데이터의 레이블 비율의 차이를 줄이기 위해 10-fold cross-validation 수행

```python
from sklearn.model_selection import StratifiedKFold

cbr = CatBoostRegressor(**cbrm_trial_params, verbose = 100)
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
```

```python
cbr_result = 0
cbr_feature_importance= 0

for train_idx,valid_idx in skf.split(trimmed_train_df,train_df['Y_Class']):
    
    x_train = trimmed_train_df.iloc[train_idx]
    x_valid = trimmed_train_df.iloc[valid_idx]
    
    y_train = train_df['Y_Quality'].iloc[train_idx]
    y_valid = train_df['Y_Quality'].iloc[valid_idx]
        
    cbr.fit(x_train,y_train,eval_set=(x_valid,y_valid),early_stopping_rounds=100)
    cbr_result += cbr.predict(trimmed_test_df)/10
    cbr_feature_importance += cbr.feature_importances_/10 # 학습할 때 변수중요도 저장됨.
```

##### 3. 분류 모델

- CatBoostClassifier를 통한 회귀 모델
- 타겟데이터의 레이블 비율의 차이를 줄이기 위해 10-fold cross-validation 수행

```python
# 시간이 너무 오래걸려 적용 못함

from sklearn.metrics import f1_score

def objective2(trial):

    cbcm_param = {
        'iterations':trial.suggest_int("iterations", 1000, 20000),
        'od_wait':trial.suggest_int('od_wait', 500, 2300),
        'learning_rate' : trial.suggest_uniform('learning_rate',0.01, 1),
        'reg_lambda': trial.suggest_uniform('reg_lambda',1e-5,100),
        'random_strength': trial.suggest_uniform('random_strength',10,50),
        'depth': trial.suggest_int('depth',1, 15),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf',1,30),
        'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations',1,15),
        'bagging_temperature' :trial.suggest_loguniform('bagging_temperature', 0.01, 100.00),
        'colsample_bylevel':trial.suggest_float('colsample_bylevel', 0.4, 1.0),
    }

    model_cbcm = CatBoostClassifier(**cbcm_param)
    model_cbcm = model_cbcm.fit(trimmed_train_df, train_df['Y_Quality'], eval_set=[(x_val, y_val)], 
                           verbose=0, early_stopping_rounds=25)

    f1_score = f1_score(y_val, model_cbrm.predict(x_val))
    return f1_score

optuna_cbcm = optuna.create_study(direction='maximize', sampler=sampler)
optuna_cbcm.optimize(objective2, n_trials = 10)
```

```python
cbc = CatBoostClassifier(verbose=100,iterations=1000)
skf = StratifiedKFold(n_splits=10,shuffle=True,random_state=100)
```

```python
cbc_result = 0
cbc_best_score = 0
cbc_feature_importance = 0

for train_idx,valid_idx in skf.split(trimmed_train_df,train_df['Y_Class']):
    
    x_train = trimmed_train_df.iloc[train_idx]
    x_valid = trimmed_train_df.iloc[valid_idx]
    y_train = train_df['Y_Class'].iloc[train_idx]
    y_valid = train_df['Y_Class'].iloc[valid_idx]    
    cbc.fit(x_train,y_train,eval_set=(x_valid,y_valid),early_stopping_rounds=100)
    cbc_result += cbc.predict_proba(trimmed_test_df)/10
    cbc_best_score += cbc.best_score_['validation']['MultiClass']/10
    cbc_feature_importance += cbc.feature_importances_/10 
```
